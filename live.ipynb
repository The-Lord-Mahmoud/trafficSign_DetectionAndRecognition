```python
from ultralytics import YOLO

# Load your custom YOLOv10 model
model = YOLO('/app/YOLOv10m/train/weights/best.pt')
```

```python
import cv2

# Initialize the webcam
cap = cv2.VideoCapture(0)  # Change the index if you have multiple cameras

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    ret, frame = cap.read()  # Read a frame from the webcam
    if not ret:
        print("Error: Could not read frame.")
        break

    # Pass the frame to the model for prediction
    img = cv2.resize(frame, (416, 416))
    results = model(img)

    # Check if results are returned and render each result on the frame
    if results:  # Make sure results is not empty
        frame = results[0].plot()  # Update frame with annotated results

    # Flip the frame horizontally after prediction
    # frame = cv2.flip(frame, 1)  # 0 for vertical flip, 1 for horizontal flip, -1 for both

    # Show the frame with the detection results
    cv2.imshow('YOLOv10 Detection', frame)

    # Break the loop if 'q' or ESC key is pressed
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q') or key == 27:  # 27 is the ESC key
        break

# Release the webcam and close all OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
