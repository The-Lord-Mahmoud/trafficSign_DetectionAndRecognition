## importing the required library
adding the pretrained model yolov10m to yolo

```python
from ultralytics import YOLO
yolo = YOLO('yolov10m')
```

Train the model, adding the data in format of yolo, yaml file contains the extentions
The output of the training is in the OUTPUT files

```python
yolo.train(data ='/kaggle/input/trafficsigns14/dataset/data.yaml',imgsz =416,epochs = 250,batch = 32)
```

```python
yolo.predict("/kaggle/input/trafficsigns14/dataset/test/images", save = True, save_txt= True)
```
